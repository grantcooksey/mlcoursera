```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import gradient_descent as gd
import math
```

```python
def next_iter(y,features,x,a):
    return x - a*gd.get_rss_partial(y,features,x)
```

```python
x = np.array([1,2,3,4])
y = np.array([2,3,4,5])**2
features, weights = gd.gen_predicted(x, 2)
```

We will use steepest descent as out direction choice

```python
n=1
x = weights
d = -gd.get_rss_partial(y, features, weights)
```

```python
steps = np.arange(0, .5, 0.00001)
len(steps)
```

```python
plot_step = []
d = -gd.get_rss_partial(y, features, x)
for step in steps:
    future= x + step*d
    rss = gd.get_rss(y,features,future)
    plot_step.append(rss)
plt.plot(steps, plot_step)
plt.show()
a = steps[plot_step.index(np.min(plot_step))]
a
```

```python
x = next_iter(y,features, x, a)
x
```

We will now plot how the minimized step size changes over time

```python
x_axis = np.array([1,2,3,4])
y_axis = np.array([2,3,4,5])**2
features, x = gd.gen_predicted(x_axis, 2)
steps = np.arange(0, 4, 0.0001)
counter = 0
step_at_iter=[]

while True:
    plot_step=[]
    counter += 1
    d = -gd.get_rss_partial(y_axis, features, x)
    for step in steps:
        future= x + step*d
        rss = gd.get_rss(y_axis,features,future)
        plot_step.append(rss)
    a = steps[plot_step.index(np.min(plot_step))]
    step_at_iter.append(a)
    
    print(counter, '|', a, '|', np.linalg.norm(d), '|', x)
        
    x = next_iter(y_axis, features, x, a)
    
    if np.linalg.norm(d)<0.01:
        break
        
plt.plot(range(counter), step_at_iter)
plt.show()
```
